## 4-4. 스타워즈 캐릭터의 성별에 따른 신장 분포

````python
df_copy["height"].describe()
# 기본적인 계산 결과 
````

````python
# 왜도 (Skewness) : 데이터가 한 쪽으로 치우쳐있는가?
# 첨도 (Turtosis) : 데이터가 얼마나 평균 근처로 몰려있는가?
````

````python
skew=df_copy["height"].skew()
turt=df_copy["height"].kurt()
````

````python
print(f"Skewness : {skew}")
print(f"Turtosis : {turt}")
````

````python
fig, ax=plt.subplots(1, 1, figsize=(5, 5), dpi=200)
_=df_copy["height"].plot.hist(bins=20, density=True, ax=ax)
_=ax.set_title("Starwars : Height")

# 위에 정규분포 그래프 추가 (비교하기 위하여)
````

````python
df_copy.loc[df_copy["sex"]=="hermaphroditic"]
# 1명이기 때문에 그냥 없애고 진행하기. (1명이기 때문에 분포를 볼 수 없다.)
````

````python
fig, ax=plt.subplots(1, 2, figsize=(10, 5), dpi=200)

# distplot : 확률분포 (히스토그램 O)
# series를 이용하여 그래프 그리기.

_=sns.distplot(df_copy.loc[df_copy["sex"]=="male", "height"], ax=ax[0])
_=sns.distplot(df_copy.loc[df_copy["sex"]=="female", "height"], ax=ax[0])
_=sns.distplot(df_copy.loc[df_copy["sex"]=="robot", "height"], ax=ax[0])

# 여러 그래프 한번에 그리는 경우, 범주 적어주기.
_=ax[0].legend(["male", "female", "robot"])

# kdeplot : 확률분포 (히스토그램 X)

_=sns.kdeplot(df_copy.loc[df_copy["sex"]=="male", "height"], ax=ax[1])
_=sns.kdeplot(df_copy.loc[df_copy["sex"]=="female", "height"], ax=ax[1])

# 여러 그래프 한번에 그리는 경우, 범주 적어주기.
_=ax[1].legend(["male", "female"])

# 키 : 일반적으로 로봇 < 여성 < 남성
````

## 4-5. 가장 가벼운 캐릭터와 가장 무거운 캐릭터

````python
df_notzero=df_copy.loc[df_copy["mass"]!=0]
````

````python
# 0을 아까 임의로 채워 넣었기 때문에, 0을 제외한 데이터에서 살펴보기.
df_notzero["mass"].unique()
````

````python
minmass=df_notzero["mass"].min()
maxmass=df_notzero["mass"].max()
````

````python
# 가장 가벼운 캐릭터
df_notzero.loc[df_notzero["mass"]==minmass]
````

````python
# 가장 무거운 캐릭터
df_notzero.loc[df_notzero["mass"]==maxmass]
````

````python
minname=df_notzero.loc[df_notzero["mass"]==minmass, "name"].values[0]
maxname=df_notzero.loc[df_notzero["mass"]==maxmass, "name"].values[0]
````

````python
print(f"가장 가벼운 캐릭터는 {minname} 이고, 몸무게는 {minmass} 이다.")
print(f"가장 무거운 캐릭터는 {maxname} 이고, 몸무게는 {maxmass} 이다.")
````
