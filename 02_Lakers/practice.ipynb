# Group By 연습하기

```python
df_copy.head()
```

```python
# date의 game_type 살펴보기
df_copy[["date", "game_type"]].groupby(["date", "game_type"]).sum()
```

```python
# 몇 번 등장했는지 세어줄 수 있다
df_copy[["date", "game_type"]].groupby(["date", "game_type"], as_index=False).size()
```

```python
# 경기 중 발생한 행동의 수
df_copy["etype"].value_counts()
```

```python
df_pivot=df_copy[["period", "etype"]].groupby(["period", "etype"], as_index=False).size()
df_pivot.head()
```

```python
df_pivot=df_pivot.pivot(index="etype", columns="period", values="size")
df_pivot
```

```python
# 각 쿼터 별 가장 많이 발생한 행동 수
df3=df_copy[["period", "etype"]].groupby(["period", "etype"], as_index=False).size()
df3.tail(15)
```

```python
# period 별 사이즈가 최대인 경우 세부 행동은 어떤 것인지 체크하기
# (1) idxmax 이용하기 - 인덱스 이용하여 찾아내기
df3.loc[df3.groupby(["period"])["size"].idxmax()]
```

```python
# (2) sort_values 처리 후 중복되는 부분 제거하기
df3.sort_values(by=["period", "size"]).drop_duplicates("period", keep="last")
```

# Group By 예제로 연습하기

```python
# 샘플 데이터
idx=['A','A','B','B','B','C','C','C','D','D','D','D','E','E','E']
col=['col1','col2','col3']
data = np.random.randint(0,9,(15,3))
df = pd.DataFrame(data=data, index=idx, columns=col).reset_index()
```

```python
df
```

```python
# 1. 인덱스 기준으로 그룹화 후 평균 구하기
# df.groupby(["index"])["col1", "col2", "col3"].mean() 
df.groupby(["index"]).mean()
```

```python
# 2. 인덱스 기준으로 그룹화 후 개수 세기
# df.groupby(["index"])["col1", "col2", "col3"].count()
df.groupby(["index"]).count()
```

```python
# 여러 개의 연산을 진행하기 위하여 사용
df.groupby(["index"]).agg(["min","max"])
```

```python
# 멀티 인덱스
idx = [['idx1','idx1','idx2','idx2','idx2'],['row1','row2','row1','row2','row3']]
col = ['col1','col2','col3']
data = np.random.randint(0,9,(5,3))
df = pd.DataFrame(data=data, index = idx, columns = col).rename_axis(index=['lv0','lv1'])
```

```python
df
```

```python
# df.groupby(level=0)["col1", "col2", "col3"].sum()
df.groupby(level=0).sum()
```

```python
df.groupby(level=0).sum().sum(axis=0)
```

```python
df.groupby(level=0).sum().sum(axis=1)
```

```python
df.groupby(level=1).sum()
```

```python
# df.groupby(level=1)['col1','col2','col3'].mean()
df.groupby(level=1).mean()
```
